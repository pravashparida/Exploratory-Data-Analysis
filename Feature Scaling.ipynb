### Feature Scaling

Feature scaling is the method to rescale the values present in the features. In feature scaling we convert the scale of different measurement into a single scale. It standardize the whole dataset in one range.

Importance of Feature Scaling

When we are dealing with independent variable or features that differ from each other in terms of range of values or units of the features, then we have to normalize/standardize the data so that the difference in range of values doesn't affect the outcome of the data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Churn_Modelling.csv')
```

```python
df.info()
```


##### What is Normalization?

Normalization is a scaling technique in which values are shifted and rescaled so that they end up ranging between 0 and 1. It is also known as Min-Max scaling.

##### What is Standardization?

Standardization is another scaling technique where the values are centered around the mean with a unit standard deviation. This means that the mean of the attribute becomes zero and the resultant distribution has a unit standard deviation.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
df.head()
```

```python
df.describe().round(2)
```

#### TASK-1 Normalization

```python
new_df = pd.DataFrame(df,columns = ['Age','Tenure'])
```

```python
new_df
```

```python
new_df.info()
```

```python
new_df['Age'] = new_df['Age'].fillna(new_df['Age'].mean())
```

```python
new_df.info()
```

```python
scaler = MinMaxScaler() # Instantiating the MinMaxScaler() function
normalized_df = scaler.fit_transform(new_df)
print(normalized_df)
```

##### Let's take another small example

```python
x_array = np.array([[2],[3],[5],[6],[6]])

scaler = MinMaxScaler()
normalized_arr = scaler.fit_transform(x_array)
print(normalized_arr)
```

##### Task-2 STANDARDIZATION

```python
scaler = StandardScaler()
standardized_df = scaler.fit_transform(new_df)
print(standardized_df)
```

```python
# Let's take another small example

x_array = np.array([[2],[3],[5],[6],[6]])

scaler = StandardScaler()
normalized_arr_ss = scaler.fit_transform(x_array)
print(normalized_arr_ss)
```
